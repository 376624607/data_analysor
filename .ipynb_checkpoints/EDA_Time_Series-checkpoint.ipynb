```python
'''
时间序列预测：ARIMA ————自回归综合移动平均

1、什么是时间序列？
    1）以固定的时间间隔收集的一组数据点
    2）不同时间之间有相关性
    3）大多数时间序列有季节性趋势
    4）也存在没有季节性趋势的时间序列：比如我们在1年中每天掷1次骰子，就不会出现数字6在
    夏季频繁出现而冬季出现少的现象

2、时间序列的稳态
    有三个标准衡量一个时间序列是否平稳时间序列数据：
    1）恒定的均值
    2）恒定的方差
    3）不随时间改变的自协方差。（自协方差是指当前时间序列数据与滞后之间数据之间的协方差）

'''
```

```python
#分析待分析时间序列数据的平稳性
plt.figure(figsize=(8,5))
plt.plot(weather_bin.Date,weather_bin.MeanTemp)
plt.title("Mean Temperature")
plt.xlabel("Date")
plt.ylabel("Mean Temp")

#从图中可以看出：气温数据出现了季节性的变化，冬暖夏凉
```

```python
timeSeries=weather_bin.loc[:,["Date","MeanTemp"]]
timeSeries.index=timeSeries.Date
ts=timeSeries.drop("Date",axis=1)
```

```python
#检验时间序列平稳状态的两种方法：
#1、画出滚动统计图：比如时间窗口为6，向后滑动，计算每个时间窗口的平均值和方差
#2、Dickey-Fuller Test:如何统计量<规定置信水平下的标准值，则是平稳的时间序列数据
```

```python
#adf-test：Augmented Dickey-Fuller检测，用于判断是时序数据是否平稳(在单变量过程中进行单位根检验，判断时间序列的相关)
#针对不同的的test版本有不同的备择假设
from statsmodels.tsa.stattools import adfuller
import warnings
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

warnings.filterwarnings("ignore")

def check_adfuller(ts):
    #Dickey-Fuller test
    result=adfuller(ts,autolag='AIC')
    print("统计量的值:",result[0])
    print("p-value",result[1])
    print("标准值:",result[4])
    
def check_mean_std(ts):
    #计算移动均值和移动方差
    rolmean=pd.rolling_mean(ts,window=6)
    rolstd=pd.rolling_std(ts,window=6)
    plt.figure(figsize=(22,10))
    orig=plt.plot(ts,color='red',label='Original')
    mean=plt.plot(rolmean,color='black',label='Rolling Mean')
    std=plt.plot(rolstd,color='green',label='Rolling Std')
    plt.xlabel("Date")
    plt.ylabel("Mean Temperature")
    plt.title("Rolling Mean & Standard Deviation")
    plt.legend()

#可视化Mean Temp 和std
check_mean_std(ts)
#检验是否平稳分布
check_adfuller(ts.MeanTemp)

'''
1、从数据可以看出：统计量的值比置信水平为0.1时的值-2.5还要大，
故不显著，不能拒绝原假设，即时序数据不平稳
2、注意pandas库中针对序列数据的rolling_mean()和rolling_std()方法
3、统计量的值通常是一个负数，负得越多，说明越拒绝原假设（是平稳的）
'''
```

```python
#使非平稳的时间序列数据平稳化
'''
时间序列数据不平稳有两个原因：
一是时变的均值，二是明显的季节性趋势(在某些时间有明显的变化)

1、解决均值随时间波动的问题：用移动平均方法
2、此处数据方差波动较小，是平稳的

'''
```

```python
#用移动平均方法实现恒定均值
moving_avg=pd.rolling_mean(ts,6)
ts_moving_arg_diff=ts-moving_avg   #注意ts_moving_arg_diff的前6个数据时Nan
ts_moving_arg_diff.dropna(inplace=True)

check_mean_std(ts_moving_arg_diff)
check_adfuller(ts_moving_arg_diff.MeanTemp)

'''
1、对时序数据求滑动平均值后，将原序列对应减去滑动均值，会发现数据变得平稳了，因为已经消除了不恒定均值的影响
2、统计量的值远远小于置信水平为0.01时的标准值-3.4392，所以显著拒绝


'''
```

```python
#差分方法使时间序列数据平稳化
plt.rcParamsDefault

ts_diff=ts-ts.shift()
plt.figure(figsize=(10,8))
plt.plot(ts_diff)
plt.title("method of difference")
plt.xlabel("Date")
plt.ylabel("Differenced Mean Temp")

#平稳化后的数据序列在0上下波动
```

```python
ts_diff.dropna(inplace=True)
check_mean_std(ts_diff)
check_adfuller(ts_diff.MeanTemp)

#P值非常小，显著拒绝原假设，即数据序列很平稳，有恒定的均值和方差
```

```python
'''
预测时间序列数据

此处选择用平稳的时间序列数据ts_diff来进行预测

方法采用ARIMA(自回归综合移动平均)：
AR(p)：自回归项，即因变量的迟滞。如果时间窗口是3的话，则用 x(t-1), x(t-2) and x(t-3)来预测x(t)
I(d)：时间序列称为平稳时间序列所需要进行的差分次数
MA(q)：移动平均，在预测时使预测误差变小

ARIMA模型的三个参数即(p,q,d)
'''
```

```python
'''
时间序列的自相关分析包括以下两个方面：
1、ACF自相关函数：度量时间序列和其迟滞版本之间的自相关性
2、PACF偏自相关函数
目的：认识时间序列的特性之后选择合适的p\q\d参数

ACF和PACF都是用于定阶，但ACF用于AR（自回归），PACF用于MA（移动平均）

从名称分析，ACF是自相关函数，并不对其他变量加以控制；而PACF是偏相关函数，就是控制其他变量后计算的自相关系数


'''
```

```python
from statsmodels.tsa.stattools import acf,pacf

#ts_diff是之前差分方法处理的平稳数据
lag_acf=acf(ts_diff,nlags=20)
lag_pacf=pacf(ts_diff,nlags=20,method='ols')
#ACF
plt.figure(figsize=(12,5))

plt.subplot(121)
plt.plot(lag_acf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(ts_diff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(ts_diff)),linestyle='--',color='gray')
plt.title("Autocorrelation Function")

#PACF
plt.subplot(122)
plt.plot(lag_pacf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(ts_diff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(ts_diff)),linestyle='--',color='gray')
plt.title("Partial Autocorrelation Function")
plt.tight_layout()

#图像分析：
'''
1、上下两条虚线是置信区间，用于确定p(自回归参数)和q(移动平均参数)
    PACF曲线第一次穿过上置信区间的点的横坐标即为p=1
    ACF曲线第一次穿过上置信区间的点的横坐标为q=1
    本文因为对ts_diff不需要进行差分了，所以d=0
    
故通过作图可以得出ARIMA模型需要的(p、q、d）三个参数
'''
```

```python
#调用ARIMA 库
from statsmodels.tsa.arima_model import ARIMA
from pandas import datetime

#拟合模型
model=ARIMA(ts,order=(1,0,1))
model_fit=model.fit(disp=0)

#预测
start_index=datetime(1944,6,25)
end_index=datetime(1945,5,31)
forecast=model_fit.predict(start=start_index,end=end_index)

#可视化
plt.figure(figsize=(10,6))
plt.plot(weather_bin.Date,weather_bin.MeanTemp,label='original')#设置label变量，方便后面legend展示
plt.plot(forecast,label='predicted')
plt.title("Time Series Forecast")
plt.xlabel("Date")
plt.ylabel("Mean Temp")
plt.legend()
```
