```python
'''
【项目04】  视频网站数据清洗整理和结论研究

作业要求：
1、数据清洗 - 去除空值
要求：创建函数
提示：fillna方法填充缺失数据，注意inplace参数

2、数据清洗 - 时间标签转化
要求：
① 将时间字段改为时间标签
② 创建函数
提示：
需要将中文日期转化为非中文日期，例如 2016年5月24日 → 2016.5.24

3、问题1 分析出不同导演电影的好评率，并筛选出TOP20
要求：
① 计算统计出不同导演的好评率，不要求创建函数
② 通过多系列柱状图，做图表可视化
提示：
① 好评率 = 好评数 / 评分人数
② 可自己设定图表风格

4、问题2 统计分析2001-2016年每年评影人数总量
要求：
① 计算统计出2001-2016年每年评影人数总量，不要求创建函数
② 通过面积图，做图表可视化，分析每年人数总量变化规律
③ 验证是否有异常值（极度异常）
④ 创建函数分析出数据外限最大最小值）
⑤ 筛选查看异常值 → 是否异常值就是每年的热门电影？
提示：
① 通过箱型图验证异常值情况
② 通过quantile(q=0.5)方法，得到四分位数
③ IQR=Q3-Q1
④ 外限：最大值区间Q3+3IQR,最小值区间Q1-3IQR （IQR=Q3-Q1）
⑤ 可自己设定图表风格

'''
```

```python
import os
os.listdir()

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

aqy_dat=pd.read_csv('爱奇艺视频数据.csv',engine='python')
print(aqy_dat.head())

aqy_dat.info()

aqy_dat.describe()

'''1、数据清洗 - 去除空值
要求：创建函数
提示：fillna方法填充缺失数据，注意inplace参数
'''
def removeNan(df):
    for i in df.columns:
        if df[i].dtype=='object':
            df[i].fillna("缺失数据",inplace=True)
        else:
            df[i].fillna(0.0,inplace=True)
    return df
new_dat=removeNan(aqy_dat)
print(new_dat)

'''2、数据清洗 - 时间标签转化
要求：
① 将时间字段改为时间标签
② 创建函数
提示：
需要将中文日期转化为非中文日期，例如 2016年5月24日 → 2016.5.24
'''
def convertDate(df,*cols):
    for i in cols:
        #注意此处必须在replace后赋值，再进行下一个replace，否则会无效果
        df[i]=df[i].str.replace('年','.')
        df[i]=df[i].str.replace('月','.')
        df[i]=df[i].str.replace('日','')
        df[i]=pd.to_datetime(df[i])
    return df
new_convert_dat=convertDate(new_dat,'数据获取日期')
print(new_convert_dat)

'''
3、问题1 分析出不同导演电影的好评率，并筛选出TOP20
要求：
① 计算统计出不同导演的好评率，不要求创建函数
② 通过多系列柱状图，做图表可视化
提示：
① 好评率 = 好评数 / 评分人数
② 可自己设定图表风格
'''

df_good_rate=new_convert_dat[['导演','好评数','评分人数']]
df_good_rate

df1=df_good_rate.groupby('导演')[['好评数','评分人数']].sum()
df1['好评率']=df1['好评数']/df1['评分人数']
df1=df1.sort_values(by='好评率',ascending=False)[:25]

#解决标签乱码问题
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
plt.figure(figsize=(12,5))
df1['好评率'].plot.bar(rot=60,label='好评率',ylim=[0.98,1])
plt.legend()

'''4、问题2 统计分析2001-2016年每年评影人数总量
要求：
① 计算统计出2001-2016年每年评影人数总量，不要求创建函数
② 通过面积图，做图表可视化，分析每年人数总量变化规律
③ 验证是否有异常值（极度异常）
④ 创建函数分析出数据外限最大最小值）
⑤ 筛选查看异常值 → 是否异常值就是每年的热门电影？
提示：
① 通过箱型图验证异常值情况
② 通过quantile(q=0.5)方法，得到四分位数
③ IQR=Q3-Q1
④ 外限：最大值区间Q3+3IQR,最小值区间Q1-3IQR （IQR=Q3-Q1）
⑤ 可自己设定图表风格
'''


#因为统计数据中有重复值存在，要先去重：比如同一部电视剧，不能将每集的人数都相加,如下所示

df2=new_convert_dat[['导演','上映年份','整理后剧名']]
df2.sort_values(by='导演',ascending=True)

df2=df2.drop_duplicates()

#筛选出不同（导演，上映年份，整理后剧名）的唯一值数据，方便后续merge操作
df3=df2.sort_values(by='导演',ascending=True)
df3=df3[df3['上映年份']!=0]
df3=df3[df3['导演']!='缺失数据']
print(df3.head())

df4=new_convert_dat.groupby('整理后剧名').sum()[['好评数','评分人数']]
df4

merge_dat=pd.merge(df3,df4,left_on='整理后剧名',right_index=True)

merge_dat

#按照上映年份分组
res_dat=merge_dat.groupby('上映年份').sum()[['评分人数','好评数']]
print(res_dat.head())

#按照评分人数作面积图
plt.figure(num=1,figsize=(12,4))
res_dat['评分人数'].loc[2000:].plot.area(alpha=0.6,label='2001-2016年评分人数统计')
plt.legend()
plt.xticks([i for i in range(2001,2017)])

#通过箱线图判断异常值
fig,axes=plt.subplots(4,4,figsize=(10,16))
start=2001
for i in range(4):
    for j in range(4):
        data=merge_dat[merge_dat['上映年份']==start]
        data[['评分人数','好评数']].boxplot(whis=3,return_type='dict',ax=axes[i,j])
        start+=1

#查看异常值

def f3(df,col):
    q1=df[col].quantile(q=0.25)
    q3=df[col].quantile(q=0.75)
    IQR=q3-q1
    tmax=q3+3*IQR
    tmin=q1-3*IQR
    return (tmin,tmax)

for i in range(2001,2017):
    df_data=merge_dat[merge_dat['上映年份']==i]
    print("{}年共有{}条数据".format(i,len(df_data)))
    theta=f3(df_data,'评分人数')[1]
    print(df_data[df_data['评分人数']>theta])
    
    
# threhold=f3(merge_dat[merge_dat['上映年份']==2016],'评分人数')
# merge_dat.loc[merge_dat['上映年份']==2016,'评分人数']>threhold[1]
```
